3. Оптимизация типов pandas

- Считайте датасет из файла Housing.csv (это данные о ценах домов) с помощью pandas
- Для каждого столбца определите оптимальный с точки зрения потребления памяти тип данных - напишите свои выводы в комментариях - 2 балла
- Поменяйте типы данных столбцов датафрейма на выбранные вами в прошлом пункте и сравните потребление памяти до и после оптимизации - 1 балл

```python
df_housing = pd.read_csv('Housing.csv')
df_housing.info()
```

```python
df_housing
```

```python
df_housing['price'] = df_housing['price'].astype('uint32')  # Значения цен положительны, поэтому используется беззнаковый тип uint32
df_housing['area'] = df_housing['area'].astype('uint16')  # Значения площадей положительны, поэтому используется беззнаковый тип uint16
df_housing['bedrooms'] = df_housing['bedrooms'].astype('uint8')  # Число спален - небольшое и положительное, можно использовать uint8
df_housing['bathrooms'] = df_housing['bathrooms'].astype('uint8')  # Число ванных комнат - небольшое и положительное, можно использовать uint8
df_housing['stories'] = df_housing['stories'].astype('uint8')  # Количество этажей - небольшое и положительное, можно использовать uint8
df_housing['mainroad'] = df_housing['mainroad'].astype('category')  # Да/нет - использование категорий
df_housing['guestroom'] = df_housing['guestroom'].astype('category')  # Да/нет - использование категорий
df_housing['basement'] = df_housing['basement'].astype('category')  # Да/нет - использование категорий
df_housing['hotwaterheating'] = df_housing['hotwaterheating'].astype('category')  # Да/нет - использование категорий
df_housing['airconditioning'] = df_housing['airconditioning'].astype('category')  # Да/нет - использование категорий
df_housing['parking'] = df_housing['parking'].astype('uint8')  # Количество парковочных мест - небольшое и положительное, можно использовать uint8
df_housing['prefarea'] = df_housing['prefarea'].astype('category')  # Да/нет - использование категорий
df_housing['furnishingstatus'] = df_housing['furnishingstatus'].astype('category')  # Ограниченное количество значений - использование категорий
```
